## Load Oil&Gas trade data

```python
# list of countries
importers=pd.read_csv('oil_importers.csv')
```

```python
fuel_intracen={}
```

```python
importers.head()
```

```python
driver.get("https://www.trademap.org/tradestat/Country_SelProductCountry_TS.aspx?nvpm=1%7c842%7c%7c%7c%7c27%7c%7c%7c2%7c1%7c1%7c1%7c2%7c1%7c2%7c1%7c1")
```

```python
for ind in importers.index[:1]:
    importer=importers.loc[ind, 'country']
    fuel_intracen[importer]={}
    
    code=str(int(importers.loc[ind, 'code']))
    code=str('0'*(3-len(code)))+ code
    
    select = Select(driver.find_element_by_id("ctl00_NavigationControl_DropDownList_Country"))
    select.select_by_value(code)
    
    check_year=driver.find_elements_by_xpath("//*[contains(text(), 'Imported value in 2019')]")
    

    all_exporters=[]
    sources={}
    
    #sort and load data
    for year in range(2001, 2019):
        element=driver.find_element_by_xpath(str("//*[contains(text(), 'Imported value in "+str(year)+"')]"))
        element.click()

        exporters=[]

        for i in range(4,10):
            elem_id=str("ctl00_PageContent_MyGridView1_ctl0"+str(i)+"_LinkButton_Country")
            val=driver.find_element_by_id(elem_id).text
            exporters.append(val)

        for i in range(10,28):
            elem_id=str("ctl00_PageContent_MyGridView1_ctl"+str(i)+"_LinkButton_Country")
            val=driver.find_element_by_id(elem_id).text
            exporters.append(val)
        exporters=[t.replace(",","") for t in exporters]    
        all_exporters.extend(exporters)
        all_exporters=list(set(all_exporters))

        ## download the table

        element=driver.find_element_by_xpath("//td[@class='tabContent']")
        text=element.text
        text=text.split('\n')
        text=text[3:]
        text=text[:-1]
        text=[t.split(" ") for t in text]
        
        target_len=min(int(np.median([len(t) for t in text])), 20)

        for t in text:    
            country=" ".join(t[:len(t)-target_len+1])
            if country not in sources.keys():
                sources[country]=[int(tt.replace(",", "")) for tt in t[len(t)-target_len+1:]]
                
    #organize the loaded data
    for year in range(2001, 2019):
        fuel_intracen[importer][year]={}
        fuel_intracen[importer][year]['sources']=list(sources.keys())
        fuel_intracen[importer][year]['values']=[sources[x][year-2001] for x in list(sources.keys())]
```

```python
import json
with open('fuel_intracen.json', 'w') as fp:
    json.dump(fuel_intracen, fp)
```
