```python
import json
```

```python
import pandas as pd
import numpy as np
```

```python
with open('intracen.json') as json_file:
    intracen=json.load(json_file)
```

#### The most growing trades between countries

```python
tradeflows=pd.DataFrame()
```

```python
for key in list(intracen.keys())[2:]:
    try:
        for t, target in enumerate(intracen[key]['export']['countries'][1:]):
            flows=pd.Series({'source':key, 'target': target}) #intracen[key]['value in 2022'] is either 1 or 0
            for year in range(2003, 2022):
                try:
                    flows.at[year] = intracen[key]['export'][str(year)][t+1]
                except: pass
            tradeflows=tradeflows.append(flows, ignore_index=True)

    except:
        pass
```

```python
flows_list = []
for key in list(intracen.keys())[2:]:
    try:
        for t, target in enumerate(intracen[key]['export']['countries'][1:]):
            flows = pd.Series({'source': key, 'target': target})
            flows_dict = {year: intracen[key]['export'].get(str(year), [None] * len(target))[t+1] for year in range(2003, 2022)}
            flows = pd.Series(flows_dict).append(flows)
            flows_list.append(flows)
    except:
        pass

tradeflows = pd.concat(flows_list, ignore_index=True)
By implementing these optimizations, you can improve the performance of the code by reducing unnecessary function calls and DataFrame operations.






Regenerate response
```
